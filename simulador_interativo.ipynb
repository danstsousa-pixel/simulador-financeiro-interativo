```python
import pandas as pd

# --- PASSO 1: COLETA DE DADOS (A Entrevista com o Usuário) ---
print("=== BEM-VINDO AO SIMULADOR DE INVESTIMENTOS ===")

tem_saldo = input("Você já tem algum dinheiro guardado para começar? (Sim/Não): ").strip().upper()

# Define o saldo inicial baseado na resposta
if tem_saldo in ['SIM', 'S', 'YES', 'Y']: 
    while True:
        try:
            capital_inicial = float(input("Qual o valor do seu saldo inicial? R$ ").replace(',', '.'))
            break
        except ValueError:
            print(" Resposta inválida! Digite apenas números (ex: 1000 ou 1000.50). Tente novamente.")
else:
    capital_inicial = 0.0

while True:
    try:
        aporte_mensal = float(input("Quanto você vai investir por mês? R$ ").replace(',', '.'))
        break
    except ValueError:
        print(" Resposta inválida! Digite apenas números para o aporte. Tente novamente.")

while True:
    try:
        meses_simulacao = int(input("Por quantos meses você pretende investir? "))
        break
    except ValueError:
        print(" Resposta inválida! Digite um número inteiro de meses (ex: 12, 24, 60).")

# --- PASSO 2: O MENU E A DEFINIÇÃO DA TAXA ---

print("\n" + "="*40)
print("Qual é o tipo de investimento?")
print("1 - Ações ou Fundos Imobiliários (FIIs) [Isento de IR]")
print("2 - Renda Fixa Tributada (Tesouro Direto ou CDB) [Com IR]")
print("3 - Renda Fixa Isenta (LCI, LCA ou Poupança) [Isento de IR]")
print("="*40)

# O VS Code vai abrir a caixinha de resposta no TOPO da tela.

while True:
    tipo_investimento = input("Olhe o menu impresso e digite sua escolha (1, 2 ou 3): ").strip()
    if tipo_investimento in ['1', '2', '3']:
        break
    else:
        print("Opção inválida! Por favor, digite apenas 1, 2 ou 3.")

# Aqui definimos a 'taxa_mensal' baseada na escolha. 
# Essa é a "mágica" (Tá feliz, Nath Acuri? KKKK) que vai fazer o dinheiro crescer ao longo do tempo.

if tipo_investimento == '1':
    taxa_mensal = 0.008  # Ações/FIIs: média de 0,8% ao mês 
    print("\n Taxa base aplicada: 0,8% ao mês (Isento)")
elif tipo_investimento == '2':
    taxa_mensal = 0.009  # Tesouro/CDB: média de 0,9% ao mês
    print("\n Taxa base aplicada: 0,9% ao mês (Tributado)")
else:
    taxa_mensal = 0.005  # Poupança/LCI: média de 0,5% ao mês
    print("\n Taxa base aplicada: 0,5% ao mês (Isento)")

# --- PASSO 3: JUROS COMPOSTOS (O Loop) ---
historico = [] # A lista vazia que vai guardar a "foto" de cada mês
saldo_atual = capital_inicial

print(" Calculando seu patrimônio financeiro...\n")

for mes in range(1, meses_simulacao + 1):
    # O dinheiro trabalha: juros sobre o saldo do mês anterior
    dividendo_do_mes = saldo_atual * taxa_mensal
    
    # O efeito bola de neve: soma tudo para virar o novo patrimônio
    patrimonio_acumulado = saldo_atual + dividendo_do_mes + aporte_mensal
    
    # Guardando a história na nossa lista 
    ## Gosto de pensar no append como "tirar uma foto" do mês atual, para depois mostrar tudo bonitinho no final. 
    # A ideia de 'anexar' as informações de cada mês é justamente para facilitar a criação da tabela depois, usando o Pandas.
    historico.append({
        'Mês': mes,
        'Saldo Inicial (R$)': round(saldo_atual, 2),
        'Aporte (R$)': round(aporte_mensal, 2),
        'Rendimento (R$)': round(dividendo_do_mes, 2),
        'Patrimônio Bruto (R$)': round(patrimonio_acumulado, 2)
    })
    
    # Atualiza o saldo para o loop girar o próximo mês corretamente
    saldo_atual = patrimonio_acumulado

# --- PASSO 4: A REGRA TRIBUTÁRIA (O "Leão" da Receita Federal) ---
total_aportado = capital_inicial + (aporte_mensal * meses_simulacao)
lucro_bruto = saldo_atual - total_aportado

aliquota_ir = 0.0

# Aplica a Tabela Regressiva APENAS se o usuário escolheu a Opção 2
if tipo_investimento == '2':
    if meses_simulacao <= 6:
        aliquota_ir = 0.225  # 22,5%
    elif meses_simulacao <= 12:
        aliquota_ir = 0.200  # 20%
    elif meses_simulacao <= 24:
        aliquota_ir = 0.175  # 17,5%
    else:
        aliquota_ir = 0.150  # 15%

valor_imposto = lucro_bruto * aliquota_ir
lucro_liquido = lucro_bruto - valor_imposto
patrimonio_liquido = total_aportado + lucro_liquido

# --- PASSO 5: EXIBIÇÃO DO RESULTADO ---
# Transforma a lista em uma tabela organizada do Pandas
df_simulador = pd.DataFrame(historico)
display(df_simulador)

print(f"\n--- RESUMO FINAL DA SIMULAÇÃO ({meses_simulacao} MESES) ---")
print(f" Total Investido (Seu esforço): R$ {total_aportado:,.2f}")
print(f" Lucro Bruto Gerado: R$ {lucro_bruto:,.2f}")

# Mostra a linha do imposto apenas se houver cobrança real
if valor_imposto > 0:
    print(f"Imposto de Renda Retido ({(aliquota_ir*100):.1f}%): - R$ {valor_imposto:,.2f}")
else:
    print("Imposto de Renda: ISENTO (R$ 0,00)")

print(f"Lucro Líquido (Dinheiro livre): R$ {lucro_liquido:,.2f}")
print(f"Patrimônio Final Líquido: R$ {patrimonio_liquido:,.2f}")
```

```python
import plotly.graph_objects as go

# 1. Preparação: Precisamos calcular o quanto o cliente tirou do bolso mês a mês
# O comando '.cumsum()' (Cumulative Sum) vai somando os valores mês a mês
df_simulador['Total Investido Acumulado'] = df_simulador['Aporte (R$)'].cumsum() 

# Se teve saldo inicial, ele entra no primeiro mês do total investido
if 'capital_inicial' in locals():
    df_simulador['Total Investido Acumulado'] += capital_inicial

# 2. Construção do Gráfico
fig = go.Figure()

# Linha 1: O seu esforço em dinheiro/investimento (Reta previsível)
fig.add_trace(go.Scatter(
    x=df_simulador['Mês'], 
    y=df_simulador['Total Investido Acumulado'],
    name='Total do Seu Bolso (Esforço)',
    mode='lines',
    line=dict(color='#7f8c8d', width=3, dash='dash') # Linha cinza tracejada
))

# Linha 2: O Patrimônio Total (A Curva da Bola de Neve)
fig.add_trace(go.Scatter(
    x=df_simulador['Mês'], 
    y=df_simulador['Patrimônio Bruto (R$)'],
    name='Patrimônio Total (Com Juros)',
    mode='lines',
    line=dict(color='#27ae60', width=4), # Linha verde forte
    fill='tonexty', # Preenche o espaço entre as duas linhas com cor
    fillcolor='rgba(39, 174, 96, 0.2)' # Um verde transparente para destacar o lucro
))

# 3. Estética do Gráfico
fig.update_layout(
    title='<b>Evolução Patrimonial: Esforço vs. Juros Compostos</b>',
    xaxis_title='Tempo (Meses)',
    yaxis_title='Valor Acumulado (R$)',
    plot_bgcolor='white',
    hovermode='x unified', # Passou o mouse, ele mostra os dois valores juntos!
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)

fig.update_xaxes(showgrid=True, gridcolor='#ecf0f1')
fig.update_yaxes(showgrid=True, gridcolor='#ecf0f1', tickprefix="R$ ")

fig.show()
```

```python
# --- PASSO 6: EXPORTAÇÃO EXCEL (Formatado por tabela) ---
import os
import pandas as pd

# 1. Preparando os dados da "Entrevista" para virar uma tabela de Resumo
if tipo_investimento == '1': nome_tipo = "Ações/FIIs (Isento)"
elif tipo_investimento == '2': nome_tipo = "Renda Fixa (Tributada)"
else: nome_tipo = "Poupança/LCI/LCA (Isento)"

resumo_dados = {
    'O que foi simulado?': [
        'Saldo Inicial', 'Aporte Mensal', 'Prazo (Meses)', 'Tipo de Investimento',
        'Taxa Mensal Aplicada', 'Total Tirado do Bolso', 'Lucro Bruto',
        'Imposto de Renda Retido', 'Lucro Líquido', 'Patrimônio Final'
    ],
    'Resultado': [
        f"R$ {capital_inicial:,.2f}", f"R$ {aporte_mensal:,.2f}", meses_simulacao, nome_tipo,
        f"{taxa_mensal*100:.2f}%", f"R$ {total_aportado:,.2f}", f"R$ {lucro_bruto:,.2f}",
        f"R$ {valor_imposto:,.2f}", f"R$ {lucro_liquido:,.2f}", f"R$ {patrimonio_liquido:,.2f}"
    ]
}

# Cria um DataFrame pequenininho só para o resumo
df_resumo = pd.DataFrame(resumo_dados)

# 2. O base de Exportação Avançada (ExcelWriter)
nome_ficheiro = 'Simulador_Completo_e_Formatado.xlsx'

# Usamos a engine 'xlsxwriter' que permite formatar o Excel por dentro do Python (Colunas mais largas, Tabelas com estilo, etc.)
with pd.ExcelWriter(nome_ficheiro, engine='xlsxwriter') as writer:
    
    # --- ABA 1: O RESUMO DA ENTREVISTA ---
    df_resumo.to_excel(writer, sheet_name='Resumo da Simulação', index=False)
    worksheet_resumo = writer.sheets['Resumo da Simulação']
    
    # Alargando as colunas para o texto não ficar espremido
    worksheet_resumo.set_column('A:A', 30) # Coluna de perguntas
    worksheet_resumo.set_column('B:B', 25) # Coluna de respostas

    # --- ABA 2: O HISTÓRICO MENSAL (Formato de Tabela) ---
    df_simulador.to_excel(writer, sheet_name='Histórico Mensal', index=False)
    worksheet_historico = writer.sheets['Histórico Mensal']
    
    # Pegamos o tamanho exato da nossa tabela (quantas linhas e colunas)
    (max_row, max_col) = df_simulador.shape
    
    # Criamos a estrutura oficial de "Tabela do Excel" com o estilo azul (Table Style Medium 9)
    col_settings = [{'header': coluna} for coluna in df_simulador.columns]
    
    worksheet_historico.add_table(0, 0, max_row, max_col - 1, {
        'columns': col_settings,
        'style': 'Table Style Medium 9',
        'name': 'TabelaEvolucaoMensal'
    })
    
    # Alargando as colunas do histórico para caber os valores em Reais
    worksheet_historico.set_column(0, max_col - 1, 18)

caminho_completo = os.path.abspath(nome_ficheiro)
print(f"SUCESSO! O Excel foi gerado.")
print(f" Caminho do arquivo: {caminho_completo}")
```
